# Select Statement: Multiplexing Channel Operations

## What is Select?

- Hình dung là `Select` giống như một cái chặn đập mà mỗi một nhánh `case` như một nhánh nước
- Nó sẽ block toàn bộ cho đến khi có 1 nhánh `case` nào đó sẵn sàng
- Khi có nhiều nhánh `case` sẵn sàng, `select` sẽ **chọn random** một nhánh case nào đó

Template cơ bản như đoạn code ở dưới

```go
select {
case msg1 := <-ch1:
    fmt.Println("Received from ch1:", msg1)
case msg2 := <-ch2:
    fmt.Println("Received from ch2:", msg2)
case ch3 <- 23:
    fmt.Println("Sent 23 to ch3")
default:
    fmt.Println("No channel operations ready")
}
```
